```python
# Series
data = ['a', 'b', 'c', 'd', 'e']

type(data)

se = pd.Series(data)
type(se)
print(se)
print(se.index)
print(se.values)
print("\n")
# slicing
se[0:3]

# Series 이름, index 이름 생성
se.name = "alphabet"
se.index.name = 'index'
print(se)
```

```python
# DF (Data Frame) = Excel : 2개 이상의 Series 데이터
data = {
    "country": ["kor", "usa", "china", "japan"],
    "rank": [1,2,3,4],
    "grade": ["A", "B", "C", "D"]
}

df =  pd.DataFrame(data)    # csv to json, json to csv, s

df
```

```python
# 데이터 설랙션 → 데이터를 불러오는 방법
# 1) df.컬럼, df['컬럼']

print(df.country)
print(df.grade)
df[['country','rank']]
```

```python
# 2) df. loc[인덱스값, 컬럼명]
print(df.loc[1])
print('\n')
print(df.loc[:])
print('\n')
print(df.loc[:, 'grade'])
df.loc[:,['grade', 'country']]
```

```python
print(df['rank'] > 2)  # boolen indexing 형태로 출력
df[df['rank'] > 2]
df.loc[df['rank'] > 2]
df.loc[df['rank'] > 2, ['grade', 'country']]
```

```python
# 삭제 (drop)
# 1) 행 데이터 삭제 방법
df.drop([3], inplace=True)
```

```python
# 2) 컬럼을 삭제
df = df.drop('rank', axis=1)
df
```

```python
df['rank'] = [1,2,3]
df
```

```python
# 기술 통계
df.info()
df.describe()
```

```python
# 데이터 정렬 → sort_index() / sort_values
df['new_rank'] = [10, 100, 30]
df.sort_values(by='new_rank', ascending=False)
```

```python
# null 데이터(NaN) 처리

# 데이터 값 변경
import numpy as np
df.loc[1, 'grade'] = np.nan

df.isnull()
df.isnull().sum()

df['grade'] = np.nan
df.fillna(value='pass', inplace=True)
```

## 연습 문제
영화 데이터 이용

```python
import pandas as pd

json_data = {
  "columns": ["Movie", "Release Year", "Audience", "Rating"],
  "index": [0, 1, 2, 3, 4, 5, 6, 7],
  "data": [
    ["Avengers", 2012, 1500, 8.8],
    ["Interstellar", 2014, 1100, 9.1],
    ["Frozen", 2013, 1020, 8.5],
    ["About Time", 2013, 950, 8.7],
    ["The Dark Knight", 2008, 1300, 9.0],
    ["Inception", 2010, 1200, 8.8],
    ["La La Land", 2016, 800, 8.6],
    ["Toy Story", 2010, 980, 8.5]
  ]
}

# JSON 데이터를 DataFrame으로 변환
df = pd.DataFrame(json_data['data'], columns=json_data['columns'])
df
```

```python
# 1) 전체 데이터 중에서 ''Moive 정보만 출력하시오.
df.Movie
df['Movie']
```

```python
# 2) 전체 데이터 중에서 'Movie','Rating' 정보를 출력하시오.
df[['Movie', 'Rating']]
```

```python
# 3) 2013년 이후에 개봉한 영화 데이터 중에서 'Movie','Rating' 정보를 출력하시오.
df[df['Release Year'] > 2013][['Movie', 'Rating']]
df.loc[df['Release Year'] > 2013,['Movie', 'Rating']]
```

```python
# 4) 주어진 계산식을 참고하여 'Recommend' Column을 추가하시오.
# Recommend = (Audience * Rating) // 100
df['Recommend'] = df['Audience'] * df['Rating'] // 100
df.sort_values(by='Recommend', ascending=False).head(5)
```

```python
# 5) 전체 데이터를 'Release Year' 기준 내림차순으로 출력하시오.
df.sort_values(by='Release Year', ascending=False)
```
